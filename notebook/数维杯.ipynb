```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('~/Downloads/A202104164840fj/美团专送数据.xlsx')
```

```python
plt.figure(figsize=(20, 12.36))
plt.xlabel('Distance/km', fontsize=20)
plt.ylabel('Time/min', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

dataT = data[data['label']=='b']
dataF = data[data['label']=='g']

T = plt.scatter(x=dataT[['distance']], y=dataT[['time']], c=dataT.label.tolist())
F = plt.scatter(x=dataF[['distance']], y=dataF[['time']], c=dataF.label.tolist())

plt.legend(handles=[T, F],labels=['T','F'],loc='best', fontsize=20)
plt.savefig("1.png")
plt.show()
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(dataF[['distance']], dataF[['time']])
reg.score(dataF[['distance']], dataF[['time']])
```
